# Basic Statistics on the Data

```python
# QUESTION: What is the average time of effect for each vaccine?  Give your answer with units
mu = tEffectPD.mean()
print(mu)

# Write the averages in a sentence. e.g., "The time of effect of vacA = xxxx minutes"
print(f"The time of effect of vacA = {mu.iloc[0]:.1f} minutes")
print(f"The time of effect of vacB = {mu.iloc[1]:.1f} minutes")
print(f"The time of effect of vacC = {mu.iloc[2]:.1f} minutes")

print(f"The time of effect of vacA = {tEffectPD["vacA"].mean():.1f} minutes")

varNames = tEffectPD.columns
print(len(varNames))
for i in range(0, len(varNames)):
    print(f"The time of effect of {varNames[i]} = {mu.iloc[i]:.1f} minutes")
    
#print(f"The time of effect of {varNames[0]} = {mu.iloc[0]:.1f} minutes")
#print(f"The time of effect of {varNames[1]} = {mu.iloc[1]:.1f} minutes")
#print(f"The time of effect of {varNames[2]} = {mu.iloc[2]:.1f} minutes")
```

```python
#QUESTION: What is the standard deviation for each vaccine?  Give your answer with units
sigma = tEffectPD.std()
print(sigma)

for i in range(0, len(varNames)):
    print(f"The standard deviation of the time of effect of {varNames[i]} = {mu.iloc[i]:.1f} +/- {sigma.iloc[i]:.2f} minutes")
```

```python
# Create a box and whisker plot that compares all 3 vaccines

#insert plt library
import matplotlib.pyplot as plt

#getting the variables in the data frame before plotting
vacA = tEffectPD["vacA"]
vacB = tEffectPD["vacB"]
vacC = tEffectPD["vacC"]

hfig, hax = plt.subplots(figsize=(5,3))
hax.boxplot([vacA, vacB, vacC])  # Create the box and whisker plot
hax.set_xticklabels(["vacA", "vacB", "vacC"]) # Create appropriate x labels
plt.ylabel("time of effect") # Create appropriate y lables
```
