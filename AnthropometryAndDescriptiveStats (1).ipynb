```python
# QUESTION: How many samples are in your data set? How would you write python code to determine # of samples?
# Print out the number of samples

hArr = data['What is your height in inches?']
wArr = data['What is your weight in pounds?']

#printing type
print(type(hArr))

# machine learning or harder data
hArr = np.array(hArr)
print(type(hArr))

print(len(hArr))

# shape is attribute the , means mutliple dimension array
print(hArr.shape)
```

```python
data.iloc[0:2,3]
```

```python
# Find the mean height and weight
# Display using print statements:
# "The average height of the class is 66.6 inches"
# "The average weight of the class is 147.0 pounds"


# keeps values that are not nan
notNans = ~np.isnan(hArr)
hArr = hArr[notNans]

notNans = ~np.isnan(wArr)
# wArr = wArr[notNans]

# method
muH = hArr.mean()
muW = wArr.mean()

print(muH)
print(muW)

# nan = not a number/null

print(f"The average height of the class is {muH} inches")
print(f"The average weight of the class is {muW} inches")
```

```python
# Find the standard deviation (method)

sigmaH = hArr.std()
sigmaW = wArr.std()

# Display the average +/- standard deviation to 1 decimal point using an f-string print statement
# : how many decimal places we want
print(f"The average height of the class is {muH:.0f} +/- {sigmaH:.0f} inches")
print(f"The average weight of the class is {muW:.0f} +/- {sigmaW:.0f} inches")
```

```python
# Plot a histogram
import matplotlib.pyplot as plt

# Histogram for height Use 5 bins
plt.hist(x = hArr, bins = 5)
plt.xlabel('height (inches)')
plt.ylabel('frequency')
```

```python
# Do histogram for weight as well
plt.hist(x = wArr, bins = 10)
plt.xlabel('weight (pounds)')
plt.ylabel('frequency')
```
